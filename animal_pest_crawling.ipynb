```python
import wikipediaapi #사용할 api 호출
import pandas as pd
```

```python
wiki = wikipediaapi.Wikipedia('Bigdata', 'en')
df = pd.read_csv('./CSV files/animal_pest_list.csv', encoding='utf-8')
for index, row in df.iterrows():
    page_py = wiki.page(row['영문'])
    if (pd.isna(row['영문'])):
        page_py = wiki.page(row['학명-수정'])
    if (page_py.exists() != True):
        page_py = wiki.page(row['학명-수정'])
    print(index + 1)
    print("page: %s" %page_py)
    print("summary: %s" %page_py.summary[0:10])
    df.at[index, '특성'] = page_py.summary[0:500]

df.to_csv('animal_pest_add_features.csv', index=False, encoding='utf-8-sig')
```
