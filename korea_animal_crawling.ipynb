```python
import wikipediaapi #사용할 api 호출
import pandas as pd
```

```python
koWiki = wikipediaapi.Wikipedia('Bigdata', 'ko')
enWiki = wikipediaapi.Wikipedia('Bigdata', 'en')
df = pd.read_csv('./CSV files/korea_animal_list.csv', encoding='utf-8')
for index, row in df.iterrows():
    page_py = enWiki.page(row['학명'])
    if (pd.isna(row['학명']) or page_py.exists() != True):
        page_py = koWiki.page(row['대표국명'])
    print(index + 1)
    print("page: %s" %page_py)
    print("summary: %s" %page_py.text[0:10])
    df.at[index, '특성'] = page_py.text[0:1500]

df.to_csv('korea_animal_add_features.csv', index=False, encoding='utf-8-sig')
```
